# Ejercicio 4

```python
def ej4(**categories):
    """
    :param categories: {
                "category1": [],
                "category2": [],
                "category3": [],
            }

    :return: dict:{
                "category_i": {
                    "values": [],
                    "gasto_total": 0,
                    "max": 0,
                    "mean": 0,
                    "rank": 0
                }
            }
    """
    final_d = {}
    ranks = []
    for k,w in categories.items():
        temp_d = {}
        temp_d['values'] = w.copy()
        temp_d['total'] = sum(w)
        temp_d['max'] = max(w)
        temp_d['mean'] = sum(w) / len(w)
        ranks.append((temp_d['total'], k))

        final_d[k] = temp_d

    ranks.sort()
    for i in range(0,len(ranks)):
        final_d[ranks[i][1]]['rank'] = len(ranks) - (i+1)
        #final_d[category]['rank']
    return final_d

print(ej4(c1=[1,2,34], c2=[2,18,21], c3=[1000]))
```
